```python
dark=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\dark.PNG")
mid=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\mid.PNG")
light=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\light.PNG")
darkgray=cv2.cvtColor(dark,cv2.COLOR_BGR2GRAY)
midgray=cv2.cvtColor(mid,cv2.COLOR_BGR2GRAY)
lightgray=cv2.cvtColor(light,cv2.COLOR_BGR2GRAY)
dark_histogram=cv2.calcHist([dark],[0],None,[256],[0,256])
mid_histogram=cv2.calcHist([mid],[0],None,[256],[0,256])
light_histogram=cv2.calcHist([light],[0],None,[256],[0,256])

plt.figure(figsize=[12,7])
plt.subplot(231);plt.imshow(darkgray,cmap="gray");plt.title("darkgray");
plt.subplot(232);plt.imshow(midgray,cmap="gray");plt.title("midgray");
plt.subplot(233);plt.imshow(lightgray,cmap="gray");plt.title("lightgray");
plt.subplot(234);plt.plot(dark_histogram);plt.title("dark_histogram");
plt.subplot(235);plt.plot(mid_histogram);plt.title("mid_histogram");
plt.subplot(236);plt.plot(light_histogram);plt.title("light_histogram");
plt.show()
```
