```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(0,170,dtype='uint8',num=180)
img=np.array([x]*50)
saturtion_channel=np.ones_like(img,dtype='uint8')*255
value_channel=np.ones_like(img,dtype='uint8')*255

hsv=cv2.merge([img,saturtion_channel,value_channel])
final_bgr_spectrum=cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
plt.imshow(final_bgr_spectrum[...,::1])
plt.show()
```

```python
image=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\004.jpg")

lower=np.array([20,50,50])
upper=np.array([35,255,255])


hsv_img=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

mask=cv2.inRange(hsv_img,lower,upper)

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(image[...,::-1]);plt.title(" image");
plt.subplot(122);plt.imshow(mask,cmap="gray");plt.title("mask1");

```

```python
image=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\004.jpg")


lower=np.array([20,50,50])
upper=np.array([35,255,255])
hsv_img=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

mask=cv2.inRange(hsv_img,lower,upper)

mask_3channel=cv2.merge([mask]*3)
result=cv2.bitwise_and(image,mask_3channel)
plt.figure(figsize=[17,7])
plt.subplot(331);plt.imshow(image[...,::-1]);plt.title(" image");
plt.subplot(332);plt.imshow(mask_3channel);plt.title("mask");
plt.subplot(333);plt.imshow(result[...,::-1]);plt.title("result");

```

```python
image=cv2.imread(r"C:\\Users\\Student\\Desktop\\pic\\004.jpg")
hsv_img=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)

# lower mask(0-10)
lower_red=np.array([0,50,50])
upper_red=np.array([10,255,255])
mask0=cv2.inRange(hsv_img,lower_red,upper_red)
# uppper mask(170-180)

lower_red=np.array([160,50,50])
upper_red=np.array([179,255,255])
mask1=cv2.inRange(hsv_img,lower_red,upper_red)

mask=mask0+mask1

plt.figure(figsize=[17,7])
plt.subplot(231);plt.imshow(image[...,::-1]);plt.title(" image");
plt.subplot(232);plt.imshow(mask0,cmap="gray");plt.title("lower mask");
plt.subplot(233);plt.imshow(mask1,cmap="gray");plt.title("upper mask");
plt.subplot(234);plt.imshow(mask,cmap="gray");plt.title("final mask");
```
